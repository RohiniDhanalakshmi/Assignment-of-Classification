```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("CKD.csv")
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
data
```

```python
data["classification_yes"].value_counts()
```

```python
data.shape
```

```python
data.columns
```

```python

indep=data[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal','pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dep=data[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'penalty':["l1", "l2", "elasticnet"],'solver':["lbfgs", "liblinear", "newton-cg", "newton-cholesky","sag", "saga"],'multi_class':["auto", "ovr", "multinomial"]}
grid=GridSearchCV(LogisticRegression(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring="f1_weighted")
grid.fit(X_train,y_train)
re=grid.cv_results_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
from sklearn.metrics import classification_report
clf=classification_report(y_pred,y_test)
print(clf)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_pred,y_test)
f1_macro
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```

```python
print("The f1_macro value for best parameter():",format(grid.best_params_),f1_macro,clf)
table=pd.DataFrame.from_dict(re)
table
```
